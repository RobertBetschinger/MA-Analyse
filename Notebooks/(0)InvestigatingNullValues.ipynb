## Wie Viele Null Values habe ich für SellerBewertung und Seller_sterne

```python
#sub_df = df[(df['seller_sterne'].isnull()) | (df['sellerbewertung'].isnull())]
sub_df = df[(df['seller_sterne'].isnull())]
# Create a new DataFrame with rows where 'sellerName' is 'Amazon', 'Amazon Warehouse', 'Amazon US', 'Amazon UK'
excluded_rows_df = sub_df[sub_df['sellerName'].isin(['amazon', 'Amazon Warehouse ', 'Amazon US', 'Amazon', 'Amazon UK'])]

# Exclude rows where 'sellerName' is 'Amazon' or 'Amazon Warehouse'
sub_df = sub_df[~sub_df['sellerName'].isin(['amazon', 'Amazon Warehouse ', 'Amazon US', 'Amazon', 'Amazon UK'])]

# Print the length of the updated sub-DataFrame and the excluded rows DataFrame
print("Length of updated sub-DataFrame:", len(sub_df))
print("Length of excluded rows DataFrame:", len(excluded_rows_df))

# Get distinct values for the 'sellerName' column in the sub-DataFrame
distinct_seller_names = sub_df['sellerName'].unique()


# Print the summarized length of these DataFrames
print("Summarized length of DataFrames:", len(sub_df) + len(excluded_rows_df))

distinct_seller_bewertungen = sub_df['sellerbewertung'].unique()
print("These are the distinct values for the seller_bewertungen", distinct_seller_bewertungen)

# Count the occurrences of each distinct value in the 'sellerbewertung' column
seller_bewertungen_counts = sub_df['sellerbewertung'].value_counts(dropna=False)

# Print the occurrence counts for each distinct value
print("Occurrence counts for distinct seller_bewertungen:")
print(seller_bewertungen_counts)

# Filter the sub_df DataFrame to keep rows with NaN values in the 'sellerbewertung' column
nan_seller_bewertung_rows = sub_df[sub_df['sellerbewertung'].isnull()]

distinct_asins = nan_seller_bewertung_rows['asin'].unique()

# Print the distinct ASINs
print("Distinct ASINs for rows with NaN values in 'seller_bewertung':")
print(distinct_asins)

# Print a few rows with NaN values in the 'sellerbewertung' column
print("Some rows with NaN values in 'seller_bewertung':")
nan_seller_bewertung_rows.head() # You can adjust the number of rows displayed by changing the argument in the head() method
```
