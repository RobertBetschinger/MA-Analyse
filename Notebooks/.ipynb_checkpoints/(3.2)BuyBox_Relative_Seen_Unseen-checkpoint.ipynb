```python
from sklearn.preprocessing import LabelEncoder

# Assuming your DataFrame is named 'df'
conditions = ['Neu', 'Gebraucht-Wieneu', 'Gebraucht-Gut', 'New', 'Gebraucht-Sehrgut',
              'Gebraucht-Akzeptabel', 'AusgezeichneterZustand(Generalüberholt)',
              'OffenerKarton-Sehrgut', 'Used-LikeNew',
              'AkzeptablerZustand(Generalüberholt)', 'Used-VeryGood', 'Used-Good',
              'OffenerKarton-Wieneu', 'Used-Acceptable', 'GuterZustand(Generalüberholt)',
              'OffenerKarton-Gut', 'OpenBox-VeryGood', 'OpenBox-LikeNew',
              'Excellentcondition(refurbished)', 'Goodcondition(refurbished)',
              'Sammlerstück-Wieneu', 'OpenBox-Good', 'Acceptablecondition(refurbished)',
              'Sammlerstück-Gut', 'Collectible-Good', 'OffenerKarton-Akzeptabel']

# Create a custom mapping for conditions
custom_mapping = {condition: i for i, condition in enumerate(conditions)}

# Ensure "Neu" and "New" have the highest ranking (lowest integer value)
custom_mapping['Neu'] = 0
custom_mapping['New'] = 0

# Apply the custom mapping to the 'condition' column
df['encoded_condition'] = df['condition'].map(custom_mapping)

# Create a LabelEncoder and fit it to the DataFrame
encoder = LabelEncoder()
encoder.fit(df['encoded_condition'])

# Transform the 'condition' column
df['encoded_condition'] = encoder.transform(df['encoded_condition'])
```

```python
## Exclude all Rows where condition isnt New and which where a reactive Crawl. No dont do that, dataset gets much smaller then
df = df[df['trigByReactive'] != True]
#df = df[df['condition'].isin(['Neu', 'New'])]

## Calculate the Price Diff For each Offer Timestamp Combination
def calculate_price_diff_and_ratio(group):
    min_price = group['price'].min()
    group['price_diff'] = group['price'] - min_price
    group['price_ratio'] = group['price'] / min_price
    return group

grouped_df = df.groupby(['asin', 'time'],group_keys=True)
#This new Dataframe isn`t grouped anymore
df = grouped_df.apply(calculate_price_diff_and_ratio)
```

```python
def encode_seller_name(seller_name):
    #amazon_variations = ['amazon', 'Amazon Warehouse', 'Amazon US', 'Amazon UK', 'Amazon']
    amazon_variations = ['amazon', 'Amazon Warehouse ']
    for index, variation in enumerate(amazon_variations, start=1):
        if seller_name == variation:
            return index
    
    return 0  # for 3P sellers

df['seller_category'] = df['sellerName'].apply(encode_seller_name)
```

```python
buy_box_counts = df['seller_category'].value_counts()
buy_box_counts
```

```python
amazon_variations = ['Amazon Warehouse ', 'Amazon US', 'Amazon UK', 'Amazon']

for variation in amazon_variations:
    count_rows = len(df[df['sellerName'] == variation])
    print(f"Number of rows sold by {variation}: {count_rows}")
```

## K-Fold Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import balanced_accuracy_score
import matplotlib.pyplot as plt

# Define the features and target
X = df[['price_diff','price_ratio','Fulfillment_type','lieferpreis','date_diff','numberOfSellers','encoded_condition','seller_category']]
y = df['buyBoxWinner']

# Define the number of folds for cross validation
n_folds = 10
kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)

# Create and train the Random Forest Classifier with cross validation
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
auc_scores = []
feature_importances = []
acc_scores = []
recall_scores = []
precision_scores = []
f1_scores = []
bal_acc_scores = []  # Added balanced accuracy scores list
for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
    # Split the data into train and test sets
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    # Fit the model and make predictions
    rfc.fit(X_train, y_train)
    y_pred_proba = rfc.predict_proba(X_test)[:, 1]  # Only take the probabilities for the positive class
    y_pred = rfc.predict(X_test)
    
    # Calculate the AUC score and feature importances for the current fold
    auc_scores.append(roc_auc_score(y_test, y_pred_proba))
    feature_importances.append(rfc.feature_importances_)
    
    # Calculate the classification report scores for the current fold
    acc_scores.append(accuracy_score(y_test, y_pred))
    recall_scores.append(classification_report(y_test, y_pred, output_dict=True)['weighted avg']['recall'])
    precision_scores.append(classification_report(y_test, y_pred, output_dict=True)['weighted avg']['precision'])
    f1_scores.append(classification_report(y_test, y_pred, output_dict=True)['weighted avg']['f1-score'])
    
    # Calculate the balanced accuracy score for the current fold
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    bal_acc_scores.append(bal_acc)  # Append balanced accuracy to the list
    
    # Print the confusion matrix using Matplotlib
    conf_matrix = confusion_matrix(y_true=y_test, y_pred=y_pred)
    conf_matrix[0,0], conf_matrix[1,1] = conf_matrix[1,1], conf_matrix[0,0]
    fig, ax = plt.subplots(figsize=(3, 3))
    ax.matshow(conf_matrix, cmap=plt.cm.Blues, alpha=0.3)
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='x-large')
    plt.xlabel('Actuals', fontsize=15)
    plt.ylabel('Predictions', fontsize=15)
    plt.title(f'Confusion Matrix - Fold {fold+1}', fontsize=15)
    plt.show()

# Print the mean AUC score across all folds
mean_auc = sum(auc_scores) / n_folds
print("Mean AUC:", mean_auc)

# Print the mean feature importances across all folds
feature_importances_df = pd.DataFrame(feature_importances, columns=X.columns)
mean_feature_importances = feature_importances_df.mean().sort_values(ascending=False)
print("Mean Feature Importances:")
print(mean_feature_importances)

# Print the mean classification report scores across all folds
mean_acc = sum(acc_scores) / n_folds
mean_recall = sum(recall_scores) / n_folds
mean_precision = sum(precision_scores) / n_folds
mean_f1 = sum(f1_scores) / n_folds
mean_bal_acc = sum(bal_acc_scores) / n_folds  # Calculate mean balanced accuracy
print("Mean AUC:", mean_auc)
print("Mean Accuracy:", mean_acc)
print("Mean Recall:", mean_recall)
print("Mean Precision:", mean_precision)
print("Mean F1-score:", mean_f1)
print("Mean Balanced Accuracy:", mean_bal_acc)  # Print mean balanced accuracy
```

```python
# Assuming the DataFrame is named 'df'
amazon_asin_values = df.loc[df['sellerName'].isin(['amazon']), 'asin'].unique()
print(f"An so vielen Asins ist Amazon als Verkäufer eingetragen:{len(amazon_asin_values)}")

# Create a new DataFrame containing only the rows with ASINs where Amazon is NOT present as a seller
df_no_amazon_asins = df[~df['asin'].isin(amazon_asin_values)]

amazon_asin_values = df_no_amazon_asins.loc[df_no_amazon_asins['sellerName'].isin(['amazon']), 'asin'].unique()
print(f"An so vielen Asins ist Amazon als Verkäufer eingetragen:{len(amazon_asin_values)}")


from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import balanced_accuracy_score
import matplotlib.pyplot as plt

# Define the features and target
X = df_no_amazon_asins[['price_diff','price_ratio','seller_sterne','sellerbewertung','Fulfillment_type','lieferpreis','date_diff','numberOfSellers','encoded_condition','seller_category']]
y = df_no_amazon_asins['buyBoxWinner']

# Define the number of folds for cross validation
n_folds = 10
kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)

# Create and train the Random Forest Classifier with cross validation
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
auc_scores = []
feature_importances = []
acc_scores = []
recall_scores = []
precision_scores = []
f1_scores = []
bal_acc_scores = []  # Added balanced accuracy scores list
for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
    # Split the data into train and test sets
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    # Fit the model and make predictions
    rfc.fit(X_train, y_train)
    y_pred_proba = rfc.predict_proba(X_test)[:, 1]  # Only take the probabilities for the positive class
    y_pred = rfc.predict(X_test)
    
    # Calculate the AUC score and feature importances for the current fold
    auc_scores.append(roc_auc_score(y_test, y_pred_proba))
    feature_importances.append(rfc.feature_importances_)
    
    # Calculate the classification report scores for the current fold
    acc_scores.append(accuracy_score(y_test, y_pred))
    recall_scores.append(classification_report(y_test, y_pred, output_dict=True)['weighted avg']['recall'])
    precision_scores.append(classification_report(y_test, y_pred, output_dict=True)['weighted avg']['precision'])
    f1_scores.append(classification_report(y_test, y_pred, output_dict=True)['weighted avg']['f1-score'])
    
    # Calculate the balanced accuracy score for the current fold
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    bal_acc_scores.append(bal_acc)  # Append balanced accuracy to the list
    
    # Print the confusion matrix using Matplotlib
    conf_matrix = confusion_matrix(y_true=y_test, y_pred=y_pred)
    conf_matrix[0,0], conf_matrix[1,1] = conf_matrix[1,1], conf_matrix[0,0]
    fig, ax = plt.subplots(figsize=(3, 3))
    ax.matshow(conf_matrix, cmap=plt.cm.Blues, alpha=0.3)
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='x-large')
    plt.xlabel('Actuals', fontsize=15)
    plt.ylabel('Predictions', fontsize=15)
    plt.title(f'Confusion Matrix - Fold {fold+1}', fontsize=15)
    plt.show()

# Print the mean AUC score across all folds
mean_auc = sum(auc_scores) / n_folds
print("Mean AUC:", mean_auc)

# Print the mean feature importances across all folds
feature_importances_df = pd.DataFrame(feature_importances, columns=X.columns)
mean_feature_importances = feature_importances_df.mean().sort_values(ascending=False)
print("Mean Feature Importances:")
print(mean_feature_importances)

# Print the mean classification report scores across all folds
mean_acc = sum(acc_scores) / n_folds
mean_recall = sum(recall_scores) / n_folds
mean_precision = sum(precision_scores) / n_folds
mean_f1 = sum(f1_scores) / n_folds
mean_bal_acc = sum(bal_acc_scores) / n_folds  # Calculate mean balanced accuracy
print("Mean AUC:", mean_auc)
print("Mean Accuracy:", mean_acc)
print("Mean Recall:", mean_recall)
print("Mean Precision:", mean_precision)
print("Mean F1-score:", mean_f1)
print("Mean Balanced Accuracy:", mean_bal_acc)  # Print mean balanced accuracy
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import balanced_accuracy_score
import matplotlib.pyplot as plt

unique_asins = df['asin'].unique()
train_asins, test_asins = train_test_split(unique_asins, test_size=0.25, random_state=56)

seen_data = df[df['asin'].isin(train_asins)]
unseen_data = df[df['asin'].isin(test_asins)]

#X_train_seen = seen_data[['is_amazon','price_diff','price_ratio','seller_sterne','sellerbewertung','Fulfillment_type','lieferpreis','date_diff','numberOfSellers']]
X_train_seen = seen_data[['price_diff','price_ratio','seller_sterne','Fulfillment_type','lieferpreis','date_diff','numberOfSellers','encoded_condition']]
y_train_seen = seen_data['buyBoxWinner']

#X_test_unseen = unseen_data[['is_amazon','price_diff','price_ratio','seller_sterne','sellerbewertung','Fulfillment_type','lieferpreis','date_diff','numberOfSellers']]
X_test_unseen = unseen_data[['price_diff','price_ratio','seller_sterne','Fulfillment_type','lieferpreis','date_diff','numberOfSellers','encoded_condition']]

y_test_unseen = unseen_data['buyBoxWinner']

rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train_seen, y_train_seen)

y_pred_proba = rfc.predict_proba(X_test_unseen)[:, 1]  # Only take the probabilities for the positive class
y_pred = rfc.predict(X_test_unseen)

conf_matrix = confusion_matrix(y_true=y_test_unseen, y_pred=y_pred)

conf_matrix[0,0], conf_matrix[1,1] = conf_matrix[1,1], conf_matrix[0,0]

fig, ax = plt.subplots(figsize=(3, 3))
ax.matshow(conf_matrix, cmap=plt.cm.Blues, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='x-large')

plt.xlabel('Actuals', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.title('Confusion Matrix', fontsize=15)
plt.show()

auc = roc_auc_score(y_test_unseen, y_pred_proba)
print("AUC:", auc)
```
