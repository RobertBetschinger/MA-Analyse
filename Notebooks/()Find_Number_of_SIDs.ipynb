```python
import pandas as pd
import numpy as np


# Open the file and read the lines
with open('../Grunddatein/ChenPaper/crawl2.txt', 'r') as file:
    lines = file.readlines()

# Initialize an empty set to store the unique third values
unique_third_values = set()

# Iterate through the lines and extract the third value
for line in lines:
    values = line.split('\t')  # Split the line by the tab delimiter
    third_value = values[2]  # Get the third value (index 2)
    unique_third_values.add(third_value)  # Add the third value to the set

# Convert the set to a list
unique_third_values_list = list(unique_third_values)

# Print the list of unique third values
print(len(unique_third_values_list))
print(unique_third_values_list)
```
